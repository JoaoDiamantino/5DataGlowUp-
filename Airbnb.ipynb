```python
# Filtrar apenas as colunas desejadas
colunas_desejadas = ['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
                     'review_scores_checkin', 'review_scores_communication', 'review_scores_location',
                     'review_scores_value']

dados_filtrados = dados[colunas_desejadas]

# Remover linhas com valores nulos nas colunas filtradas
dados_filtrados_sem_nulos = dados_filtrados.dropna()
```

```python
import statsmodels.api as sm

data = dados_filtrados_sem_nulos.copy()

# Coluna de Intercepto
data['intercepto'] = 1

# Especifique as variáveis independentes (X) e a variável dependente (y)
X = data[[ 'review_scores_accuracy', 'review_scores_cleanliness',
                     'review_scores_checkin', 'review_scores_communication', 'review_scores_location',
                     'review_scores_value', 'intercepto']]
y = data['review_scores_rating']

# Ajuste o modelo de regressão linear múltipla
modelo = sm.OLS(y, X).fit()

# Obtenha os coeficientes da regressão
coeficientes = modelo.params

# Calcule o impacto relativo em termos percentuais
impacto_percentual = (coeficientes / coeficientes['intercepto']) * 100
```

```python
# Coeficientes excluindo o intercepto
coeficientes = {
    'review_scores_accuracy': 2.542398,
    'review_scores_cleanliness': 2.348750,
    'review_scores_checkin': 0.699798,
    'review_scores_communication': 1.779182,
    'review_scores_location': 0.385392,
    'review_scores_value': 2.647219
}

# Calcular a soma dos valores absolutos dos coeficientes excluindo o intercepto
soma_absoluta = sum(abs(coeficiente) for coeficiente in coeficientes.values())

# Calcular a porcentagem relativa de cada coeficiente em relação à soma dos valores absolutos excluindo o intercepto
pesos_em_porcentagem = {segmento: (abs(coeficiente) / soma_absoluta) * 100 for segmento, coeficiente in coeficientes.items()}

# Criar um DataFrame a partir do dicionário
tabela_pesos = pd.DataFrame(list(pesos_em_porcentagem.items()), columns=['Segmento', 'Porcentagem Relativa'])

# Ordenar a tabela em ordem decrescente pela coluna 'Porcentagem Relativa'
tabela_pesos = tabela_pesos.sort_values(by='Porcentagem Relativa', ascending=False)

# Mostrar a tabela
print(tabela_pesos)

```
